````sos
# xQTL Model Training Tutorial

## Overview
This tutorial demonstrates the new configurable xQTL model training pipeline.

### Key Features:
- **Two YAML configuration files** for data and model parameters
- **Clean command-line interface** using argparse
- **Four CatBoost models** for xQTL prediction
- **Chromosome-based train/test splits**

### Files Created:
1. `data_config.yml` - Data paths and parameters
2. `model_config.yml` - Model hyperparameters
3. `simple_trainer.py` - Main training script
````

````sos
# Display configuration files
import yaml

print("=== Data Configuration ===")
with open('data_config.yml', 'r') as f:
    data_config = yaml.safe_load(f)
    
for key, value in data_config.items():
    print(f"{key}: {value}")

print("\n=== Model Configuration ===")
with open('model_config.yml', 'r') as f:
    model_config = yaml.safe_load(f)
    
for key, value in model_config.items():
    print(f"{key}: {value}")
````
